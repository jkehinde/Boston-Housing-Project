We can initially make some assumption for each feature:

1) House that have moe room (RM has a higher value) are usually more expensive (higher MDEV). In short an increase in RM will lead to an increase MDEV(Price)

2) Neigheborhoods with a higher proportion of lower class workers (higher LSTAT) will have houses that are worth less. So an increase in LSAT will lead to a decrease in MDEV(price) of houses. The reason being that lower class workers cannot afford expensive house.

3) Finally, Neighborhoods with more student to teacher ratio (higher PTRATIO) will be have a lower MDEV(price) of houses. This is because the lack of teachers to student may indicate there is less tax income in the area (as a result of low income workers).
